```python
import pandas as pd
import eurostat

# Import data using the Eurostat API
data = eurostat.get_data_df('TOUR_CE_OMN12')

#Drop columns'freq','unit'
data.drop(columns=['freq','unit'], errors='ignore', inplace=True)

# Check the structure of the data
print(data.head())
```

```python
#add column Region to label geographic regions (NUTS codes).
df = pd.DataFrame(data)
df['Region'] = df['geo\TIME_PERIOD'].apply(lambda x: 'NUT1' if len(x) == 3 else ('NUT2' if len(x) == 4 else ('Country' if len(x) == 2 else '[EU27_2020]')))
#put column "geo" besides column "Region" 
col_order = ['geo\TIME_PERIOD', 'Region'] + [col for col in df.columns if col not in ['geo\TIME_PERIOD', 'Region']]
df = df[col_order]

#Drop row where 'c_resid' and 'month' value is total
df = df[df['c_resid'] != 'TOTAL']
df = df[df['month'] != 'TOTAL']

df.columns.values[df.columns.get_loc('geo\TIME_PERIOD')] = 'geo'
#output_file = 'processed_data.xlsx'
#df.to_excel(output_file, index=False)
df
```

```python
long_df = pd.melt(df, id_vars=['geo', 'Region', 'indic_to', 'c_resid', 'month'], value_vars=['2018', '2019', '2020', '2021', '2022', '2023', '2024'], var_name='Year', value_name='Value') 
long_df
```

```python
missing_values = long_df[(long_df['Year'] == 2018) & (long_df['Value'].isna())]
missing_values
```

```python
#Check null value amount for each column
null_counts = long_df.isnull().sum()
print(null_counts)
```

```python
#Check null value amount for each column
null_counts = df.isnull().sum()
print(null_counts)

#Since the amount of 2018-2023 null value is small, so we set them to 0.
columns_to_fill = ['2018', '2019', '2020', '2021', '2022', '2023', '2024']
df[columns_to_fill] = df[columns_to_fill].fillna(0)
df
```

```python
#Summarize data by month, guest residence, or region
summary_df = df.groupby(['month', 'c_resid', 'indic_to', 'Region']).agg({
    '2018': 'sum',
    '2019': 'sum',
    '2020': 'sum',
    '2021': 'sum',
    '2022': 'sum',
    '2023': 'sum'
}).reset_index()
summary_df
```

```python
#Transfer format of summary_df
long_summary_df = summary_df.melt(id_vars=['month', 'c_resid', 'indic_to', 'Region'],
                                    var_name='Year', value_name='Value')

long_summary_df
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px
# Convert the month column to a category type and specify the order of the categories
month_order = [f"M{str(i).zfill(2)}" for i in range(1, 13)]
summary_df['month'] = pd.Categorical(summary_df['month'], categories=month_order, ordered=True)

# melt datasheet to adapt to plotly's long-form formatting
df_melted = summary_df.melt(id_vars=['month', 'c_resid', 'indic_to', 'Region'], value_vars=['2018', '2019', '2020', '2021', '2022', '2023'], var_name='Year', value_name='Value')

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='region-dropdown',
        options=[{'label': region, 'value': region} for region in summary_df['Region'].unique()],
        value=summary_df['Region'].unique()[0]
    ),
    dcc.Dropdown(
        id='indic_to-dropdown',
        options=[{'label': indic_to, 'value': indic_to} for indic_to in summary_df['indic_to'].unique()],
        value=summary_df['indic_to'].unique()[0]
    ),
    dcc.Dropdown(
        id='c_resid-dropdown',
        options=[{'label': 'All', 'value': 'All'}] + [{'label': c_resid, 'value': c_resid} for c_resid in summary_df['c_resid'].unique()],
        value='All'
    ),
    dcc.Graph(id='line-chart')
])

@app.callback(
    Output('line-chart', 'figure'),
    [Input('region-dropdown', 'value'),
     Input('indic_to-dropdown', 'value'),
     Input('c_resid-dropdown', 'value')]
)
def update_chart(selected_region, selected_indic_to, selected_c_resid):
    if selected_c_resid == 'All':
        filtered_df = df_melted[
            (df_melted['Region'] == selected_region) &
            (df_melted['indic_to'] == selected_indic_to)
        ].groupby(['month', 'Year']).sum().reset_index()
    else:
        filtered_df = df_melted[
            (df_melted['Region'] == selected_region) &
            (df_melted['indic_to'] == selected_indic_to) &
            (df_melted['c_resid'] == selected_c_resid)
        ]
    
    fig = px.line(
        filtered_df, x='month', y='Value', color='Year', 
        title=f"Line Chart for {selected_region}, {selected_indic_to}, {selected_c_resid}"
    )
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
